# Задание 1

## Межфазные потоки Al-содержащих компонент на поверхности твердого алюминия

```python
x0 = [1, 1, 1, 1, 1]
task1 = Task1(x0)

plt.xlabel(u'1 / T, 1 / K')
plt.ylabel(u'ln(G), кмоль / (м^2 * сек)')
T1 = 623.15
T2 = 923.15
T = np.arange(T1, T2)
Y = lambda idx : np.array([np.log(task1.interfacial_G(t)[idx]) for t in T])
plt.plot(1 / T, Y(0), label = u'AlCl')
plt.plot(1 / T, Y(1), label = u'AlCl2')
plt.plot(1 / T, Y(2), label = u'AlCl3')
plt.legend()
plt.show()
```

## Cкорость испарения источника алюминия

```python
plt.xlabel('1 / T, 1 / K')
plt.ylabel('ln(V), нм / сек')
T1 = 623.15
T2 = 923.15
T = np.arange(T1, T2)
lnV = np.array([np.log(task1.calc_V(t)) for t in T])
plt.plot(1 / T, lnV)
plt.show()
```

# Задание 3

## Межфазные потоки компонент AlCl3 и GaCl на ростовой поверхности

```python
x0 = [0.000000, 0.053926, 1481.463522, 21.127906, 6.441113, 0.008038]
task3 = Task3(x0)

plt.xlabel(u'Доля AlCl3')
plt.ylabel(u'G, кмоль / (м^2 * сек)')
AC3 = np.arange(0, 1, 0.01)
Y0 = lambda idx : np.array([(task3.interfacial_G(x_g, 0)[idx]) for x_g in AC3])
plt.plot(AC3, Y0(0), label = u'AlCl3(Pg(H2) = 0)')
plt.plot(AC3, Y0(1), label = u'GaCl(Pg(H2) = 0)')
Y1 = lambda idx : np.array([(task3.interfacial_G(x_g, 1)[idx]) for x_g in AC3])
plt.plot(AC3, Y1(0), label = u'AlCl3(Pg(H2) = 9847)')
plt.plot(AC3, Y1(1), label = u'GaCl(Pg(H2) = 9847)')
plt.legend(loc='center left',  bbox_to_anchor=(1, 0.5))
plt.show()
```

## Cкорость роста слоя AlGaN
