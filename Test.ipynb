```python
import matplotlib.pyplot as plt
import numpy as np
import math

x = np.linspace(0, 1)
# y = 4 * x ** 3 + 6 * x ** 2 - 20 * x - 13
def y(x):
    return (-x * np.log2(x)) - (1 - x)  *  np.log2(1 - x)
plt.plot(x, y(x))
plt.grid(True)
plt.axhline(0, color='black')
plt.axvline(0, color='black')
```
